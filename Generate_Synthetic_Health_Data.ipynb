```python
df_temp = df[['Height ','Weight ','BMI','How many glassess of water do you drink a day?','How many alcohol drinks do you have a week?']].copy()
df_temp.head(5)
```

```python
print(df['How many glassess of water do you drink a day?'].unique())
print(df['How many alcohol drinks do you have a week?'].unique())
print(df['Do you eat 3 or more servings of whole grains per day?'].unique())
print(df['Do you eat fats like nuts or fish 2 or more times a week'].unique())
print(df['Do you usually eat butter, cheese or drink full cream milk or curd 2 or more times per week'].unique())
print(df['Do you eat sweet treats on most of the days of the week?'].unique())
print(df['Do you eat deep fired or junk food on most of the days of the week?'].unique())
print(df['Do you eat 5 or more fruitsand vegetables per day?'].unique())
```

```python
waterPerDay = {'0-5 glass':0, '6-9 glass':1, '10+ glass':2}
alcPerWeek = {'1-4 servings/week ':1, '1-4 servings/fortnightly':2, 'none ':0, '>4 servings/week ':3}
df_temp['How many glassess of water do you drink a day?'] = df_temp['How many glassess of water do you drink a day?'].map(waterPerDay)
df_temp['How many alcohol drinks do you have a week?'] = df_temp['How many alcohol drinks do you have a week?'].map(alcPerWeek)
```

```python
df_temp.dtypes
```

```python
df_temp['How many glassess of water do you drink a day?'] = df_temp['How many glassess of water do you drink a day?'].astype('category')
df_temp['How many alcohol drinks do you have a week?'] = df_temp['How many alcohol drinks do you have a week?'].astype('category')
```
